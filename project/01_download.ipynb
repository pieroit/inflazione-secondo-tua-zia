```python
import pandas as pd
import numpy as np
import plotly.express as px
import requests
from bs4 import BeautifulSoup
import itertools
from tqdm import tqdm
```

```python

url = 'https://osservaprezzi.mise.gov.it/prezzi/livelli/beni-e-servizi-di-largo-consumo/archivio-rilevazioni-beni-e-servizi-di-largo-consumo?ANNO=2021&MESE=9&f%5BPROVINCIA%5D=Roma&f%5BTIPO_RECORD_MISE%5D=altri_alim&submit=Applica'
df = pd.read_html(url)[0]
df.sort_values('Descrizione Prodotto').reset_index(drop=True)
```

```python
form_options = {}
for input_name in [ 'f[PROVINCIA]', 'f[TIPO_RECORD_MISE]', 'ANNO', 'MESE' ]:
    
    response = requests.get('https://osservaprezzi.mise.gov.it/prezzi/livelli/beni-e-servizi-di-largo-consumo/archivio-rilevazioni-beni-e-servizi-di-largo-consumo?ANNO=2021&MESE=7&f%5BPROVINCIA%5D=Roma&f%5BTIPO_RECORD_MISE%5D=altri_alim&submit=Applica')

    soup = BeautifulSoup(response.text)
    select = soup.find(attrs={'name':input_name})
    
    options = []
    for option in select.find_all('option'):
        options.append( option["value"].replace(' ', '%20') )
    form_options[input_name] = options
    
form_options
```

```python

def get_table(url):

    df = pd.read_html(url)[0]
    df = df.sort_values('Descrizione Prodotto').reset_index(drop=True)
    return df   

all_tables = []
form_options['MESE'] = [9] # less data! TODO: all

for anno in form_options['ANNO']:
    for mese in form_options['MESE']:
        for provincia in tqdm(form_options['f[PROVINCIA]']):
            for tipologia in form_options['f[TIPO_RECORD_MISE]']:
                
                url = 'https://osservaprezzi.mise.gov.it/prezzi/livelli/beni-e-servizi-di-largo-consumo/archivio-rilevazioni-beni-e-servizi-di-largo-consumo?'
                url += f'ANNO={anno}&MESE={mese}&f%5BPROVINCIA%5D={provincia}&f%5BTIPO_RECORD_MISE%5D={tipologia}&submit=Applica'
    
                table = get_table(url)
                
                table['anno'] = anno
                table['mese'] = mese
                table['provincia'] = provincia.replace('%20', ' ')
                table['tipologia'] = tipologia
                #display(table)
        
                all_tables.append(table)
        
prezzi = pd.concat(all_tables)
prezzi
```

```python
prezzi.to_csv('prezzi.csv', index=False)
```
